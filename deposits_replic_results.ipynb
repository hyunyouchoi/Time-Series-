# 1. SAV

```python
df_sav = pd.read_csv('df_sav.csv')
```

```python
df_sav.shape
```

```python
df_sav.columns
```

```python
df_sav_first = df_sav.groupby('accountNo').first().reset_index()
```

```python
df_sav_first.shape
```

```python
#filedate and origination date
df_sav_first['fileDate'] = pd.to_datetime(df_sav_first['fileDate'])
df_sav_first['originationDate'] = pd.to_datetime(df_sav_first['originationDate'])

df_sav_first['fileYM'] = df_sav_first['fileDate'].apply(lambda x: datetime.strftime(x, "%Y-%m"))
df_sav_first['originYM'] = df_sav_first['originationDate'].apply(lambda x: datetime.strftime(x, "%Y-%m"))
```

```python
df_sav_first.groupby('fileDate').count()
```

```python
df_sav_first_notnew = df_sav_first[df_sav_first['fileYM'] != df_sav_first['originYM']]
```

```python
df_sav_first_notnew.shape
```

```python
df_sav_first_notnew_count = df_sav_first_notnew.groupby('fileDate').count().reset_index()
df_sav_first_notnew_sum = df_sav_first_notnew.groupby('fileDate').sum().reset_index()
```

```python
df_sav_first_notnew_count.to_csv('df_sav_first_notnew_count.csv')
df_sav_first_notnew_sum.to_csv('df_sav_first_notnew_sum.csv')
```
